```python
#importing packages
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv("../Datasets/NB.csv")
y = pd.read_csv("../Datasets/IR - 7.csv")
```

```python
print(X.shape)
print(y.shape)
```

```python
y.columns
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X))
# Random shuffle training data
X_train.sample(frac=1)

X_test = pd.DataFrame(scaler.transform(y))
```

```python
from keras.layers import Input, Dropout
from keras.layers.core import Dense 
from keras.models import Model, Sequential, load_model
from keras import regularizers
```

```python
import tensorflow as tf
from numpy.random import seed
seed(10)
tf.random.set_seed(10)
act_func = 'elu'
```

```python
input_dim = X_train.shape[1]
encoding_dim = 32
```

```python
input_layer = Input(shape=(input_dim, ))

encoder = Dense(encoding_dim, activation="tanh", 
                activity_regularizer=regularizers.l1(10e-5))(input_layer)
encoder = Dense(int(encoding_dim / 2), activation="relu")(encoder)

decoder = Dense(int(encoding_dim / 2), activation='tanh')(encoder)
decoder = Dense(input_dim, activation='relu')(decoder)

autoencoder = Model(inputs=input_layer, outputs=decoder)
```

```python
nb_epoch = 100
batch_size = 1000

autoencoder.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
autoencoder.summary()
```

```python
NUM_EPOCHS=100
BATCH_SIZE=1000
history=autoencoder.fit(X_test,X_test, 
                  batch_size=BATCH_SIZE, 
                  epochs=NUM_EPOCHS,
                  validation_data=(X_train, X_train),
                  verbose = 1,
                  shuffle = False)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], 'b', label='Training loss')
plt.plot(history.history['val_loss'], 'r', label='Validation loss')
plt.legend(loc='upper right')
plt.xlabel('Epochs')
plt.ylabel('Loss, [mse]')
plt.ylim([0,.05])
plt.show()
```

```python
X_pred_train = autoencoder.predict(X_train)
```

```python
import seaborn as sns

scored = np.mean(np.abs(X_pred_train-X_train), axis = 1)
plt.figure()
g = sns.distplot(scored, bins = 25, kde= True);
#g.set(xlim=(0.05,0.10), ylim=(0, 1))
plt.show()
```

```python
TH = 0.026

scored_train = pd.DataFrame()
scored_train['Loss_mae'] = np.mean(np.abs(X_pred_train-X_train), axis = 1)
scored_train['Threshold'] = TH
scored_train['Anomaly'] = scored_train['Loss_mae'] > scored_train['Threshold']

scored_train
```
