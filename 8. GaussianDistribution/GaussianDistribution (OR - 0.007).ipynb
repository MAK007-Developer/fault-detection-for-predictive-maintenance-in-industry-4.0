```python
#importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Importing the dataset
X = pd.read_csv('../Datasets/NB.csv')
X['Fault'] = 0

y = pd.read_csv('../Datasets/OR - 7.csv')
y['Fault'] = 1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(X, test_size = 0.2, shuffle=False, random_state = 0)

y_train, y_test = train_test_split(y, test_size = 0.2, shuffle=False, random_state = 0)
```

```python
train = X_train.append(y_train)
train = train.reset_index(drop=True)
train
```

```python
test = X_test.append(y_test)
test = test.reset_index(drop=True)
test
```

```python
from sklearn.covariance import EllipticEnvelope

df_class0 = train.loc[train['Fault'] == 0, 'DE']
df_class1 = train.loc[train['Fault'] == 1, 'DE']

outliers_fraction = 0.01

envelope =  EllipticEnvelope(contamination = outliers_fraction) 
train_data = df_class0.values.reshape(-1,1)
envelope.fit(train_data)
df_class0 = pd.DataFrame(df_class0)
df_class0['deviation'] = envelope.decision_function(train_data)
df_class0['anomaly'] = envelope.predict(train_data)

envelope =  EllipticEnvelope(contamination = outliers_fraction) 
train_data = df_class1.values.reshape(-1,1)
envelope.fit(train_data)
df_class1 = pd.DataFrame(df_class1)
df_class1['deviation'] = envelope.decision_function(train_data)
df_class1['anomaly'] = envelope.predict(train_data)

df_class = pd.concat([df_class0, df_class1])

train['anomaly'] = df_class['anomaly']

train_anomalies = train[train['anomaly'] == -1]

f, (ax4) = plt.subplots(figsize=(18, 6))
ax4.scatter(train_anomalies.index, train_anomalies.DE, label='anomaly', color='red', s=10)
ax4.plot(train.index, train.DE, label='DE');
plt.xlim((0,len(train.index)))

plt.title('Gaussian Distribution')
plt.legend();
plt.show();
```

```python
from sklearn.covariance import EllipticEnvelope

df_class = test['DE']

test_data = df_class.values.reshape(-1,1)
df_class = pd.DataFrame(df_class)
df_class['deviation'] = envelope.decision_function(test_data)
df_class['anomaly'] = envelope.predict(test_data)

test['anomaly'] = df_class['anomaly']

test_anomalies = test[test['anomaly'] == -1]

f, (ax4) = plt.subplots(figsize=(18, 6))
ax4.scatter(test_anomalies.index, test_anomalies.DE, label='anomaly', color='red', s=10)
ax4.plot(test.index, test.DE, label='DE');
plt.xlim((0,len(test.index)))

plt.title('Gaussian Distribution')
plt.legend();
plt.show();
```

```python
from sklearn.metrics import accuracy_score
score = 100 * accuracy_score(test['anomaly'], test['Fault'])
print("Accuracy: {:.2f}%".format(score))
```
